```python
import pandas as pd
import re 

df = pd.read_csv("combined_file.csv")

# انتقال اولین جمله تا اولین نقطه به ستون جدید question و بقیه متن به ستون answer
df['question'] = df['text'].apply(lambda x: x.split('.')[0])
df['answer'] = df['text'].apply(lambda x: '.'.join(x.split('.')[1:]).strip())

# به‌روزرسانی ستون text تا مشابه answer شود
df['context'] = df['answer']
df['context'] = df['context'].apply(lambda x: re.sub(r'\[.*?\]', '', x))
df['answer'] = df['answer'].apply(lambda x: re.sub(r'\[.*?\]', '', x))

# حذف عبارت - ویکی خمینی از ستون title
df['title'] = df['title'].str.replace(' - ویکی امام خمینی', '')

# انتخاب فقط ستون‌های مورد نیاز
df = df[['title', 'question', 'answer', 'context']]

# ذخیره فایل اصلاح شده
df.to_csv("crawl_wiki_khomeini.csv", index=False)
```
