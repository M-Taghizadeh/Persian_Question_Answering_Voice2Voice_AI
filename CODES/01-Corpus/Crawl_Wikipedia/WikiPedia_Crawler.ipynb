```python
from collections import deque
from tqdm import tqdm

max_pages = 50000
urls = ['ایران']

def scraped_data(urls, max_pages):
    scraped_dataset = {
        'url': [],
        'title': [],
        'summary': [],
        'text': [],
    }

    visited_urls = set()
    queue = deque(urls)
    count = 0

    # Initialize tqdm progress bar
    progress_bar = tqdm(total=max_pages, desc="Scraping pages")

    while queue and count < max_pages:
        url = queue.popleft()
        if url in visited_urls:
            continue

        visited_urls.add(url)
        page = wiki_wiki.page(url)

        try:
            # Add page data to the dataset
            scraped_dataset['url'].append(page.fullurl)
            scraped_dataset['title'].append(page.title)
            scraped_dataset['summary'].append(page.summary)
            scraped_dataset['text'].append(page.text)

            # Add new links to the queue
            for link in page.links:
                if link not in visited_urls and count < max_pages:
                    queue.append(link)

            count += 1
            # Update progress bar
            progress_bar.update(1)
        except Exception as e: pass

    # Close the progress bar
    progress_bar.close()

    return scraped_dataset

# Example usage
scraped_dataset = scraped_data(urls, max_pages=max_pages)
```
