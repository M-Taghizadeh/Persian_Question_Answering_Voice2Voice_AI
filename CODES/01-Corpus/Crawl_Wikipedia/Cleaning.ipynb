```python
# اضافه کردن ستون جدید برای نشان دادن اینکه آیا صفحه حاوی محتوای ضد رژیم است یا خیر
df['is_anti_regime'] = df['summary'].apply(lambda x: contains_keywords(x, keywords))

# فیلتر کردن صفحات حاوی محتوای ضد رژیم
filtered_df = df[df['is_anti_regime'] == 0]
filtered_df = filtered_df[['title', 'summary', 'url']]

# ذخیره داده‌های فیلتر شده در یک فایل CSV جدید
filtered_df.to_csv("filtered_iran_wikipedia_pages.csv", index=False)
```
