```python
train_df = pd.read_csv("/content/drive/MyDrive/QA_Babaei/train.csv")
test_df = pd.read_csv("/content/drive/MyDrive/QA_Babaei/test.csv")
```

```python
for i in range(len(train_df['title'])):
  if train_df['answer'][i]=="نامشخص":
    train_df['answer'][i] = str(train_df['context'][i]).split('.')[0]
    train_df['answer_en'][i] = str(train_df['context_en']).split('.')[0]
```

```python
for i in range(len(test_df['title'])):
  if test_df['answer'][i]=="نامشخص":
    test_df['answer'][i] = str(test_df['context'][i]).split('.')[0]
    test_df['answer_en'][i] = str(test_df['context_en']).split('.')[0]
```

```python
import pandas as pd

# خواندن فایل CSV
df = pd.read_csv('train_46486.csv')

# فیلتر کردن ردیف‌هایی که طول ستون context آنها کمتر از 3 نیستند
df_filtered = df[df['context'].apply(lambda x: len(str(x)) >= 3)]

# ذخیره دیتاست جدید (در صورت نیاز)
df_filtered.to_csv('filtered_train_46486.csv', index=False, encoding='utf-8-sig')

# نمایش دیتاست جدید
print(df_filtered)
```
